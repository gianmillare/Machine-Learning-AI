# Kmeans Activity - More Clusters

```python
# Step 1: Import Dependencies

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
from sklearn.datasets import make_blobs
from sklearn.cluster import KMeans
```

```python
# Step 2: Get the Data

data, _ = make_blobs(n_samples=500, centers=6, cluster_std=0.60, random_state=0)
plt.scatter(data[:,0], data[:,1])
plt.show()
```

```python
# Step 3: Create the KMeans model
kmeans = KMeans(n_clusters=12)
```

```python
# Step 4: Train (fit) the data to the model
kmeans.fit(data)
```

```python
# Step 4: Use the model to classify, make predictions
predicted_clusters=kmeans.predict(data)
plt.scatter(data[:,0], data[:,1], c=predicted_clusters,s=50,cmap='viridis')
plt.show()
```

# Optional: visualize the centers

```python
centers=kmeans.cluster_centers_
labels=kmeans.labels_
```

```python
print(centers)
print("------------" * 4)
print(labels)
```

```python
plt.scatter(data[:,0], data[:,1], c=labels, s=50,cmap='viridis')
plt.scatter(centers[:,0], centers[:,1], c='red', s=150, alpha=0.5)

plt.show()
```
