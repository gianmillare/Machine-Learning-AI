# Moons - Deep Learning

```python
# Step 1: Import Dependencies

%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import sklearn
import sklearn.datasets
from numpy.random import seed
seed(1)
```

### Data Preprocessing

```python
# Step 2: Get the data, set X and y

X, y = sklearn.datasets.make_moons(200, noise=0.05)
plt.scatter(X[:, 0], X[:, 1], c=y)
```

```python
print(X[:5])
print("--------" * 10)
print(X[:5, 0])
print("--------" * 10)
print(X[:5, 1])
```

```python
# Step 3: Train test split the X and y

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=1)
```

```python
# Step 4: StandardScaler to scale the data and fit the training features (X)

from sklearn.preprocessing import StandardScaler

X_scaler = StandardScaler().fit(X_train)
```

```python
# Step 5: because the data was scaled, transform the X_train and X_test data

X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Step 6: Because the y values are 1's and 0's, we one hot encode the data as readible to the model

from keras.utils import to_categorical

# One Hot Encoding
y_train_categorical = to_categorical(y_train)
y_test_categorical = to_categorical(y_test)
```

### Create the Neural Network

```python
# Step 7: Create a Neural Network
from keras.models import Sequential
from keras.layers import Dense

model = Sequential()
model.add(Dense(units=6, activation='relu', input_dim=2))
model.add(Dense(units=2, activation='softmax'))
```

```python
model.summary()
```

```python
# Step 8: Compile the Neural Network Model

model.compile(optimizer='adam', loss='categorical_crossentropy', metrics=['accuracy'])
```

```python
# Step 9: Fit the neural network to the training (scaled) model

model.fit(X_train_scaled, y_train_categorical, epochs=500, shuffle=True, verbose=2)
```

### Create the Deep Learning Neural Network

```python
# Step 10: create the dlnn with more layers

deep_model = Sequential()
deep_model.add(Dense(units=6, activation="relu", input_dim=2))
deep_model.add(Dense(units=6, activation="relu"))
deep_model.add(Dense(units=2, activation="softmax"))
```

```python
deep_model.summary()
```

```python
# Step 11: Compile the deep learning model

deep_model.compile(optimizer='adam', loss='categorical_crossentropy', metrics='accuracy')
```

```python
# Step 12: Fit the deep learning model using scaled training data

deep_model.fit(X_train_scaled, y_train_categorical, epochs=300, shuffle=True, verbose=2)
```

```python
# Step 13: Compare the 2 models using testing data

model_loss, model_accuracy = model.evaluate(X_test_scaled, y_test_categorical, verbose=2)
print(f"Normal NN - Loss: {model_loss}, Accuracy: {model_accuracy}")

deep_loss, deep_accuracy = deep_model.evaluate(X_test_scaled, y_test_categorical, verbose=2)
print(f"Deep Learning NN - Loss: {deep_loss}, Accuracy: {deep_accuracy}")
```

### Key takeaway is that the DLNN is more accurate and had less of a loss compared to the Normal NN. Also, it took less epochs for the DLNN to acheive a higher accuracy.
